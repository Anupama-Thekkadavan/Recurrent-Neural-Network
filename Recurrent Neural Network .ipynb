```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
int_to_char = dict((i, c) for i, c in enumerate(alphabets))
int_to_char
```

```python
char_to_int = dict((c, i) for i, c in enumerate (alphabets))
```

```python
char_to_int
```

```python
seq_length = 1
x = []
y = []

for i in range(len(alphabets) - 1):
    seq_in = alphabets[i:i+seq_length]
    seq_out = alphabets[i+seq_length]
    print(seq_in, '->', seq_out)
```

```python
seq_length = 1
x = []
y = []

for i in range(len(alphabets) - 1):
    seq_in = alphabets[i:i+seq_length]
    seq_out = alphabets[i+seq_length]
    print(seq_in, '->', seq_out)
    x.append([char_to_int[char] for char in seq_in])
    y.append(char_to_int[seq_out])
```

```python
import numpy as np
```

```python
X = np.reshape(x, (len(x), seq_length, 1))
```

<b>Build the model<b>

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
#Create the object of RNN
model = Sequential()

#Add the LSTM layer
layer1 = LSTM(32, input_shape = (1, 1))
model.add(layer1)

#ADD the output layer
layer2 = Dense(26, activation='softmax')
model.add(layer2)
```

```python
model.summary()
```

```python
from keras.utils import plot_model
```

```python
plot_model (model, show_layer_activations=True, show_layer_names=True, show_shapes=True)
```

<b>Compile the model<b>

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics =['accuracy'])
```

<b>Train the model<b>

```python
model.fit(X, y, epochs=1000, batch_size=1)
```

<b>Evaluate the model<b>

```python
model.evaluate(X, y, batch_size=1)
```
